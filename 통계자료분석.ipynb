```python
kdif=[0]
for i in range(1,304):
    value=kd['confirmed'][i]-kd['confirmed'][i-1]
    kdif.append(value)

kd['diff']=kdif

d_max=kd['diff'].max()
d_min=kd['diff'].min()
c_max=kd['confirmed'].max()
c_min=kd['confirmed'].min()
param1=[]
param3=[]
for i in range(0,304):
    value2=((kd['diff'][i]-d_min)/(d_max-d_min))
    value3=((kd['confirmed'][i]-c_min)/(c_max-c_min))
    param1.append(value2)
    param3.append(value3)
kd['param1']=param1
kd['param3']=param3
kd
```

```python
sdif=[0]
for i in range(1,10):
    value=-(sb['BSI'][i]-sb['BSI'][i-1])
    sdif.append(value)

sb['diff']=sdif

d_max=sb['diff'].max()
d_min=sb['diff'].min()
param2=[]
for i in range(0,10):
    value2=(sb['diff'][i]-d_min)/(d_max-d_min)
    param2.append(value2)
    
sb['param2']=param2
sb
```

```python
new_index=[]
for i in range(0,10):
    new_index.append(sb.index[i].replace("2020. ",""))
sb.index=new_index
sb
```

```python
param2=[]
for i in range(0,304):
    if kd['date'][i].split('-')[1]=='11':
        param2.append(0)
    else:
        for j in range(0,10):            
            if kd['date'][i].split('-')[1]==sb.index[j]:
                param2.append(sb['param2'][j])
kd['param2']=param2
kd
```
